```python
#task_1
#a
import pandas as pd
movie_performance = pd.read_csv('C:\\Users\\Katia\\Documents\\dc_marvel_movie_performance.csv')
movie_performance.head()
```

```python
#b
columns_to_remove = ['MCU', 'Phase', 'Franchise']
movie_performance = movie_performance.drop(columns=columns_to_remove)
```

```python
print(movie_performance.columns)
```

```python
#c
character_family=movie_performance.drop_duplicates(subset=['Character Family'], keep='first', inplace = False)
```

```python
character_family.head()
```

```python
character_family = character_family['Character Family'].sort_values()
print(character_family)
```

```python
#d 
longest_row_index = movie_performance['Length'].idxmax()
longest_row = movie_performance.loc[longest_row_index]

print(longest_row)
```

```python
#e
mean_per_break_even = movie_performance.groupby('Break Even')['Rotten Tomatoes Critic Score'].mean()
print(mean_per_break_even)
```

```python
#f
movie_performance['Recomended'] = movie_performance['Rotten Tomatoes Critic Score'].apply(lambda x: 'Strongly Yes' if x >= 90 else ('Yes' if (x >= 60 and x < 90) else ('No' if (x >= 40 and x < 60) else 'Strongly No')))
movie_performance.head()
```

```python
movie_performance.value_counts("Recomended")
```

```python
movie_performance.value_counts("Recomended", normalize = True)
```

```python
#g
character_family_agg = movie_performance.pivot_table(index='Character Family', columns='Break Even', aggfunc='size', fill_value=0).reset_index()
character_family_agg['Success Rate'] = character_family_agg['Success']/ (character_family_agg['Flop'] + character_family_agg['Success'])
character_family_agg['Success Rate'] = character_family_agg['Success Rate'].apply(lambda x: f"{x:.2%}")
print(character_family_agg)
```

```python
#task_2
#a
top_5_earnings = movie_performance['Gross to Budget'].nlargest(5)
top_5_earnings_rows = movie_performance[movie_performance['Gross to Budget'].isin(top_5_earnings)]
top_5_earnings_rows.head()
```

```python
films = top_5_earnings_rows['Film'].tolist()
print(films)
```

```python
gross_to_budget = top_5_earnings_rows['Gross to Budget'].tolist()
print(gross_to_budget)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.bar(films, gross_to_budget)
ax.set_ylabel('earnings')
ax.set_title('films earnings')
plt.xticks(rotation=45)
plt.show()
```
